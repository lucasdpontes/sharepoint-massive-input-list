```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import Chrome
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
# Parâmetros da biblioteca Selenium
sharepoint_lista_url = 'LINK DE ACESSO À LISTA'
TEMPO_DE_ESPERA = 60
driver = Chrome()
driver.get(sharepoint_lista_url)
actions = ActionChains(driver)
driver.maximize_window()

# Fonte de dados (pode ser adaptado para outros formatos, como CSV)
arquivo = pd.read_excel('Fonte de dados.xlsx')

colunas_verificadas = arquivo.columns[arquivo.columns != 'Cadastrado']

for linha in range(0,len(arquivo)):

    # Verifica se há alguma coluna vazia
    if arquivo.iloc[linha][colunas_verificadas].isnull().any():
        
        arquivo.loc[linha, 'Cadastrado'] = 'Não'
        arquivo.to_excel('Fonte de dados.xlsx',index=False)
    
    # Inicia a criação dos registros, caso a coluna 'Cadastrado' esteja vazia
    elif arquivo.loc[linha, 'Cadastrado'] not in ['Sim','Não','Pular']:

        coluna_a = arquivo['Coluna A'][linha]
        coluna_b = arquivo['Coluna B'][linha]
        coluna_c = arquivo['Coluna C'][linha]
        coluna_d = arquivo['Coluna D'][linha]
        coluna_e = arquivo['Coluna E'][linha]
        coluna_f = arquivo['Coluna F'][linha]
        coluna_g = arquivo['Coluna G'][linha]
        coluna_h = arquivo['Coluna H'][linha]
        coluna_i = arquivo['Coluna I'][linha]
        coluna_j = arquivo['Coluna J'][linha]

        # Se o código parar nesse bloco com muita frequência, utilizar a linha 38 ao invés da 39
        button_selector = 'button[name="Novo"]'
#        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button = WebDriverWait(driver, TEMPO_DE_ESPERA).until(EC.element_to_be_clickable((By.CSS_SELECTOR, button_selector)))
        button.click()
        
        #Campo "Coluna A"
        button_selector = 'div#displayView-displayDiv-COLUNA A'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        element = driver.switch_to.active_element
        element.send_keys(coluna_a)
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna B"
        button_selector = 'div[title="'+coluna_b+'"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna C"
        button_selector = 'div[title="'+coluna_c+'"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna D"
        button_selector = 'div[title="'+coluna_d+'"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna E"
        button_selector = 'div#displayView-displayDiv-COLUNA E'
        element = driver.switch_to.active_element
        element.send_keys(coluna_e)
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna F"
        button_selector = 'div[title="'+coluna_f+'"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna G"
        button_selector = 'div[title="'+coluna_g+'"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna H"
        button_selector = 'div#displayView-displayDiv-COLUNA H'
        element = driver.switch_to.active_element
        element.send_keys(coluna_h)
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna I"
        element = driver.switch_to.active_element
        element.send_keys(coluna_i)
        element.send_keys(Keys.ENTER)
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
        
        #Campo "Coluna J"
        button_selector = 'div[title="'+coluna_j+'"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        element = driver.switch_to.active_element
        element.send_keys(Keys.TAB)
         
        button_selector = 'button[name="Salvar"]'
        button = WebDriverWait(driver,TEMPO_DE_ESPERA).until(EC.presence_of_element_located((By.CSS_SELECTOR, button_selector)))
        button.click()
        
        # Atualiza a coluna de controle e salva o arquivo
        arquivo.loc[linha, 'Cadastrado'] = 'Sim'
        arquivo.to_excel('Fonte de dados.xlsx',index=False)
        del coluna_a,coluna_b,coluna_c,coluna_d,coluna_e,coluna_f,coluna_g,coluna_h,coluna_i,coluna_j
```
